# Question 1: What is the difference between a function and a method in Python?

Answer:

In Python, both function and method are used to perform specific tasks, but they are not exactly the same.

A function is an independent block of code that is defined using the def keyword and can be called directly by its name. It does not belong to any class.

A method, on the other hand, is a function that is defined inside a class and is used to perform operations related to objects of that class. A method is always called using an object.

For example, print() is a function, but append() is a method because it is used with list objects like list.append().

# Question 2: Explain the concept of function arguments and parameters in Python.

Answer:

When we define a function, the variables written inside the parentheses are called parameters. They act as placeholders for the values that will be passed later.

When we call the function and pass actual values to it, those values are called arguments.

```python
def add(a, b):   # a and b are parameters
    result = a + b
    return result  # Function should return the result

my_sum = add(5, 3)   # 5 and 3 are arguments
print(f"The sum is: {my_sum}")
```

# Question 3: What is the purpose of the return statement in a Python function?

Answer:

The return statement is used to send a result back from a function to the place where it was called.

Without the return statement, a function performs its task but does not give back any value. When return is used, the function stops executing and sends the specified value to the caller.

```python
def square(n):
    return n * n
    print("This line will not be executed")

n = 10
result = square(n)
print(f"square of the {n} is {result}")
```

# Question 4: Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
num = int(input("Enter a number: "))
Factorial = 1
i = 1

while i <= num:
    Factorial = Factorial * i
    i += 1

print(f"Factorial of {num} is {Factorial}")
```

# Question 5: What is Object-Oriented Programming (OOP)?

Answer:

Object-Oriented Programming (OOP) is a programming approach in which we organize code using objects and classes instead of only functions.

In OOP, data and functions are combined into a single unit called an object. This approach makes programs more organized, reusable, and easier to manage, especially for large projects.

Main concepts of OOP include encapsulation, abstraction, inheritance, and polymorphism.

# Question 6: What is a class and an object in OOP?

Answer:

A class is like a blueprint or template that defines the properties and behaviors of something.

An object is a real instance created from that class. It contains actual data and can use the methods defined in the class.

```python
# example
class student(): # student is a class
    def __init__(self, name, age):
        self.name = name
        self.age = age
student1 = student("Parwez", 25) # student1 is object
print(student1.name)
print(student1.age)
```

# Question 7: What is the difference between abstraction and encapsulation?

Answer:

Both abstraction and encapsulation are important concepts in OOP, but they are slightly different.

**Abstraction** means hiding complex implementation details and showing only the important features to the user. It focuses on what an object does rather than how it does it.

**Encapsulation** means wrapping data and methods together inside a class and restricting direct access to some data. It helps in protecting data from accidental modification.

# Question 8: Write Python code to create a class Person with attributes name and age, and a method display that prints these attributes.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)

# Creating object
p1 = Person("Parwez", 25)
p1.display() # call from display method
```

# Question 9: What is inheritance in OOP? Provide a simple example in Python.

Answer:

Inheritance is a feature in OOP where one class can use the properties and methods of another class. It helps in reusing code and reducing repetition.

The class that gives properties is called the parent class, and the class that receives them is called the child class.

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog() # called Dog class (child claas)
d.speak() # recieve from parent class
d.bark() # will receive from the called class Dog (child class)
```

# Question 10: What is a lambda function in Python? Provide an example.
Answer:

A lambda function is a small anonymous function written in a single line using the lambda keyword. It does not require a function name.

It is usually used for short operations.

```python

square = lambda x: x * x
print(square(5))
```
