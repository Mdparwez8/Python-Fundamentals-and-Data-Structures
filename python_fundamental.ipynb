# Question 1 : What is mutability in Python?
Mutable objects can be modified after creation.
Examples: list, dictionary, set

Immutable objects cannot be changed after creation.
Examples: int, float, string, tuple

For example, elements inside a list can be changed, but elements inside a tuple cannot.

```python
#Question 2: Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.

# Creating a list
my_list = [10, 20, 30]
my_list[0] = 100   # Changing element
print("Modified List:", my_list)

# Creating a tuple
my_tuple = (10, 20, 30)
my_tuple[0] = 100   # This will give an error
print("Tuple:", my_tuple)
```

# Question 3: What are logical operators in Python?

Answer:

Logical operators are used to combine conditional statements.

Python has three logical operators:

and – Returns True if both conditions are true

or – Returns True if at least one condition is true

not – Reverses the result (True becomes False, False becomes True)

```python
a = 10
b = 5

print(a > 5 and b > 2)  # True
print(a > 5 or b > 10)  # True
print(not(a > 5))       # False
```

# Question 7: What is the difference between for and while loops?

Answer:

For Loop - Used when number of iterations is known

While Loop - Used when number of iterations is unknown

```python
# examples
# For loop
for i in range(5): # range = 0 to (n-1)
    print(i) # will print 0, 1, 2, 3, 4

# While loop
i = 0
while i < 5: # while printing it will add the value by 1 in each case and stops after getting value 5.
    print(i)
    i += 1
```

# Question 8: Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial is:", factorial)
```

Question 9: Why might you use a set instead of a list in Python?

Answer:

A set is used instead of a list when:

we want to store unique elements only

Order of elements does not matter

```python
numbers = {1, 9, 2, 3, 3, 4, 4, 5, 3} # sets does not print the repeated elements
print(numbers)  # Output: {1, 2, 3, 4, 5, 9} # orders does not matter
```
